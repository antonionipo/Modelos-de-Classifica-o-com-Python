O banco que solicitou o modelo de limite de crédito incluiu novas variáveis no dataset como escolaridade, gênero e Estado do cliente.
Faça as transformações necessárias nessas variáveis Dummies e tire o modelo OLS até que o modelo fique ajustado.

```python
import pandas as pd
import seaborn as sns
import numpy as np
import statsmodels.formula.api as smf
```

```python
df = pd.read_excel('Limite_Credito_Dummy.xlsx')
df.head()
```

```python
df = pd.get_dummies(df, columns=['Escolaridade', 'Gênero', 'Região'], prefix=['Escolaridade', 'Gênero', 'Região'])
df.head()
```

```python
df.corr()
```

```python
function = 'LimitedoChequeEspecial~Salario+Região_F+Região_M - 1'
```

```python
model = smf.ols(formula=function, data=df).fit()
print(model.summary())
```
