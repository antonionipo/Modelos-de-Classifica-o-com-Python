A DNC solicitou a você que com os dados de Investimento, investindo em Marketing e Gastos Administrativos seja possível prever o lucro da escola.
Faça a análise desses dados e utilize algum método de seleção de variáveis para adicionar ao modelo somente as variáveis explicativas que tenham o p-value até 5%.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.api as sm
```

```python
df = pd.read_excel('Exercicio_Lucro.xlsx')
df.head()
```

```python
df.corr()
```

```python
sns.pairplot(df)
```

```python
function = 'Lucro ~ Investimento + Investimento_Mkt + Gastos_Administrativos'
model = sm.OLS.from_formula(function, data=df).fit()
print(model.summary())
```

```python
function = 'Lucro ~ Investimento + Investimento_Mkt' # Retirar Gastos Administrativos
model = sm.OLS.from_formula(function, data=df).fit()
print(model.summary())
```
