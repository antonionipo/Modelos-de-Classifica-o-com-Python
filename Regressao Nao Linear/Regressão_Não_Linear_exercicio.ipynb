```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
from sklearn import metrics
from sklearn.tree import export_graphviz
import pydot
```

```python
features = pd.read_excel('house.xlsx')
features.head(5)
```

```python
features.describe()
```

```python
features.info()
```

```python
features = features.dropna()
```

```python
features = pd.get_dummies(features)
features.head()
```

```python
labels = np.array(features['SalePrice'])
features= features.drop('SalePrice', axis = 1)
feature_list = list(features.columns)
features = np.array(features)
```

```python
features_train, features_test, labels_train, labels_test = train_test_split(features, labels, test_size = 0.25, random_state = 42)
```

```python
rf = RandomForestRegressor(n_estimators = 1000, random_state = 42)
rf.fit(features_train, labels_train);
ada = AdaBoostRegressor(n_estimators = 1000, random_state = 42)
ada.fit(features_train, labels_train);
gbr = GradientBoostingRegressor(n_estimators = 1000, random_state = 42)
gbr.fit(features_train, labels_train);

rf_pred = rf.predict(features_test)
rf_e = abs(rf_pred - labels_test)
rf_rsqrt = rf.score(features, labels)
ada_pred = ada.predict(features_test)
ada_e = abs(ada_pred - labels_test)
ada_rsqrt = ada.score(features, labels)
gbr_pred = gbr.predict(features_test)
gbr_e = abs(gbr_pred - labels_test)
gbr_rsqrt = gbr.score(features, labels)

print('Random Forest')
print('R-squared:', round(rf_rsqrt, 2))
print('Mean Absolute Error:', round(metrics.mean_absolute_error(labels_test, rf_pred), 2))
print('Mean Squared Error:', round(metrics.mean_squared_error(labels_test, rf_pred), 2))
print('Ada Boost')
print('R-squared:', round(ada_rsqrt, 2))
print('Mean Absolute Error:', round(metrics.mean_absolute_error(labels_test, ada_pred), 2))
print('Mean Squared Error:', round(metrics.mean_squared_error(labels_test, ada_pred), 2))
print('Gradient Boost')
print('R-squared:', round(gbr_rsqrt, 2))
print('Mean Absolute Error:', round(metrics.mean_absolute_error(labels_test, gbr_pred), 2))
print('Mean Squared Error:', round(metrics.mean_squared_error(labels_test, gbr_pred), 2))
```

```python
tree = rf.estimators_[5]
export_graphviz(tree, out_file = 'tree.dot', feature_names = feature_list, rounded = True, precision = 1)
(graph, ) = pydot.graph_from_dot_file('tree.dot')
graph.write_png('tree.png')
```

```python
for model in [rf, ada, gbr]:
  importances = list(model.feature_importances_)
  feature_importances = [(feature, round(importance, 2)) for feature, importance in zip(feature_list, importances)]
  feature_importances = sorted(feature_importances, key = lambda x: x[1], reverse = True)
  print(model)
  [print('Feature: {:20} Importance {}'.format(*pair)) for pair in feature_importances]
  print('\n')
```
